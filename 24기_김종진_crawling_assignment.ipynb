```python
import requests

url = 'https://www.musinsa.com/app/codimap/lists?style_type=&tag_no=&brand=&display_cnt=60&list_kind=big&sort=comment_cnt&page=1'
res = requests.get(url, headers={"User-Agent" : "Mozilla/5.0"})

print(res) # 응답 객체
html = res.text # HTML 텍스트
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.find('span', class_='style-list-information__text').get_text()
```

```python
style_list = soup.find_all('span', class_='style-list-information__text')
for i in style_list :
  print(i.get_text())
```

```python
soup.find('strong', class_='style-list-information__title').get_text()
```

```python
style_title = soup.find_all('strong', class_='style-list-information__title')
for i in style_title :
  print(i.get_text())
```

```python
soup.find('span', class_='post-information__text').get_text()
```

```python
post_information = soup.find_all('span', class_='post-information__text')
x = 0
date_list = []
for i in range(60) :
  date_list.append(post_information[x])
  x += 3

y = 1
view_list = []
for i in range(60) :
  view_list.append(post_information[y])
  y += 3

z = 2
comment_list = []
for i in range(60) :
  comment_list.append(post_information[z])
  z += 3
```

```python
import pandas as pd

result_df = pd.DataFrame(columns = ['codimap_category', 'codimap_title', 'codimap_date', 'views', 'comment_numbers'])
```

```python
append_list = [style_list, style_title, date_list, view_list, comment_list]
column = 0
for i in append_list :
  empty_list = []
  for x in i :
    empty_list.append(x.get_text())
  result_df.iloc[:,column] = empty_list
  column +=1

result_df['views'] = result_df['views'].str.replace('조회', '').str.strip()
result_df['comment_numbers'] = result_df['comment_numbers'].str.replace('댓글', '').str.strip()
```
